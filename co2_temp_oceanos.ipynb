## CO2 - Histórico de emissões globais de CO2 (1750-2020)
Publicado por Ian Tiseo , 9 de maio de 2023
(https://www.statista.com/statistics/264699/worldwide-co2-emissions/)

As emissões de dióxido de carbono liberadas pela combustão global de combustíveis fósseis e processos industriais tiveram um aumento dramático desde o surgimento da revolução industrial. As emissões de dióxido de carbono (CO2) começaram a aumentar mais acentuadamente a partir da década de 1950 e, em 2000, atingiram 25,23 bilhões de toneladas métricas de CO2. As emissões dispararam 32% entre 2000 e 2010 e, em 2020, totalizaram 34,81 bilhões de toneladas métricas. 

O que é dióxido de carbono?
O dióxido de carbono é um gás natural incolor que é liberado depois que pessoas e animais inalam oxigênio. É um gás de efeito estufa, o que significa que absorve e libera radiação térmica que, por sua vez, cria o *“efeito estufa”*. Além de outros gases de efeito estufa, o dióxido de carbono também é um dos principais contribuintes para a capacidade da Terra de manter uma temperatura habitável. Sem dióxido de carbono e outros gases de efeito estufa, a Terra seria muito fria para se viver. No entanto, embora o dióxido de carbono por si só não seja um gás prejudicial, sua **abundância** é o que causa as mudanças climáticas. O aumento do uso de __eletricidade__ , __transporte__ e __desmatamento__ na sociedade humana resultaram no aumento das emissões de dióxido de carbono. Isso, por sua vez, tem resultado num aumento na temperatura da Terra. As consequências disto são desertificação, inundações e ondas de calor em pontos variados da superfície da terra.

## Anomalia da temperatura

O que é uma anomalia? 

Uma anomalia é um comportamento **não habitual** de um processo. Por exemplo, para calcular a anomalia da temperatura da superfície do mar relativa ao mês de Dezembro de 1997, temos de subtrair a média da temperatura de todos os meses de Dezembro de um período de referência (por ex., dos últimos vinte anos) ao mês em questão (isto é, Dezembro de 1997). O resultado é expresso em valores positivos ou negativos, em que zero é o valor “normal”. Se o resultado for positivo, o parâmetro está anormalmente alto. Se o resultado for negativo, o parâmetro está anormalmente baixo (https://www.esa.int/SPECIALS/Eduspace_Weather_PT/SEMIFPYOBFG_0.html#:~:text=Por%20exemplo%2C%20para%20calcular%20a,%C3%A9%2C%20Dezembro%20de%201997 Agência Espacial Européia). 

Dessa forma, as anomalias são desvios da média em relação à média. Quando ocorrem variações sazonais, muitas vezes é util normalizar os dados. A normalização é obtida dividindo-se a anomalia pelo desvio-padrão climatológico. Dessa forma, oferece mais informação sobre a magnitude das anomalias porque as influências de dispersões foram removidas (https://iridl.ldeo.columbia.edu/dochelp/StatTutorial/Climatologies/index.html#Intro). Este pode ser um outro tópico para análise de séries temporais. 

No caso específico do gráfico apresentado aqui, o período base de cálculo da temperatura média da água da superfícial dos oceanos considerou o período de 1901 - 2000. (https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/global/time-series/globe/ocean/ann/12/1880-2022)

```python
# Global Ocean Temperature Anomalies, January-December
# Units: Degrees Celsius
# Base Period: 1901-2000

# Os valores presentes no data frame "teta_oceanos" indica a diferença da temperatura para os respectivos anos de 1880 a 2022. 
# obs.: para efeito de ajustar o plot do gráfico com emissão de CO2, será considerado o período de 1880 a 2022, 

teta_oceanos = pd.read_csv('/content/temperatura_oceanos.txt') 
teta_oceanos
```

```python
# extração ('drop') dos anos 2021 e 2022
teta_oceanos.drop([141,142], axis = 0, inplace = True) 
```

```python
teta_oceanos
```

```python
# Anomalia anual na temperatura superficial dos oceanos de 1880 a 2020 (temperatura em graus Celsius)
# https://www.statista.com/statistics/224893/land-and-ocean-temperature-anomalies-based-on-temperature-departure/

sns.lineplot(data=teta_oceanos, x = 'Year', y = 'Value')
```

```python
# Emissão Global de CO2 a partir de combustíveis fósseis.

co_dois = pd.read_excel('/content/global-historical-co2-emissions-from-fossil-fuels-and-industry-1750-2020.xlsx', sheet_name = 'Data1') #, usecols = [1,4], skiprows=134) 

# A quantidade de CO2 está em bilhões de toneladas. 

# Fonte de DDs: https://ourworldindata.org/co2-emissions
```

```python
co_dois
```

```python
# plot da qt// de CO2 emitido a partir de combustíveis fósseis.

sns.lineplot(data=co_dois , x='Year', y='CO2')
```

```python
# juntando ("join') os 2 data frames (<teta_oceanos> e <CO2>): 
#   * Oscilação temperatura dos oceanos
#   * Qt// CO2
 
df_union = pd.merge(teta_oceanos, co_dois, how = 'inner', on = 'Year')
```

```python
df_union
```

```python
# plot do gráfico com eixo y à esquerda (anomalia da temperatura)
ax1 = sns.lineplot(
    x='Year', 
    y='Value', 
    data=df_union, 
    sort=False, 
    color='Blue'
)
ax1.set_ylabel('Oscilação da temperatura superficial oceânica')
ax1.set_xlabel('Ano')
ax1.legend(['Temperatura'], loc="upper left")

###############################

# plot do gráfico com eixo y à direita (ton de CO2)
ax2 = ax1.twinx()
sns.lineplot(
    x='Year', 
    y='CO2', 
    data=df_union, 
    color='red', 
    alpha=0.5, 
    ax = ax2       
)
# ax2.grid(b=False) # turn off grid #2
ax2.set_ylabel('Emissão de CO2 em bilhões de ton')
ax2.legend(['CO2'], loc="upper center")
```

# Conclusões

A observação do gráfico mostra que por volta de 1950, época de ocorrência da "Revolução Industrial", há um ponto de inflexão da curva de emissão de CO2. O crescimento da emissão de CO2 passa a ocorrer a uma taxa maior de incremento. Simultaneo a isto, há um aumento das anomalias da temperatura superficial dos oceanos. As anomalias da temperatura são positivas, isto é, ha um aumento da temperatura observada em relação à média do período de 1901 a 2000. Para plotar os dois gráficos e a quantidade de anos permanecer igual, excluí os anos de 2021 e 2022 da temperatura.
